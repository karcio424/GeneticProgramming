```python
import os
import numpy as np
import math

def load_data(file_name):
    with open(file_name, 'r') as file:
        num_variables = int(file.readline().strip())
        function_expression = file.readline().strip()
        num_intervals = int(file.readline().strip())
        intervals = [list(map(float, line.split())) for line in file.readlines()]

    return num_variables, function_expression, num_intervals, intervals

def save_results(file_name, results, num_variables):
    with open(file_name, 'w') as file:
        if num_variables == 1:
            file.write("1 100 -5 5 201\n")  # Dodaj pierwszą linię dla funkcji jednej zmiennej
        elif num_variables == 2:
                    file.write("2 100 -5 5 10201\n")  # Dodaj pierwszą linię dla funkcji dwóch zmiennych

        for result in results:
            if len(result) == 2:
                file.write(f'{result[0]:.4f}\t{result[1]:.4f}\n')
            elif len(result) == 3:
                file.write(f'{result[0]:.4f}\t{result[1]:.4f}\t{result[2]:.4f}\n')

def single_variable_function(x, expression):
    return eval(expression)

def multi_variable_function(x, y, expression):
    return eval(expression)

def calculate_step(interval, num_variables):
    start, end = interval
    factor = 100 if num_variables == 2 else 200
    return (end - start) / (factor)

def process_data(file_prefix, x_range_start, x_range_end):
    input_folder = 'filesAutomated/function_data'
    output_folder = 'filesAutomated/data'

    for i in range(x_range_start, x_range_end + 1):
        file_name = os.path.join(input_folder, f'{file_prefix}{i}.dat')

        if not os.path.exists(file_name):
            print(f"Plik {file_name} nie istnieje. Pomijam.")
            continue

        num_variables, function_expression, num_intervals, intervals = load_data(file_name)

        for j, interval in enumerate(intervals, start=1):
            start, end = interval
            step = calculate_step(interval, num_variables)

            results = []

            if num_variables == 1:
                results = [(x, single_variable_function(x, function_expression)) for x in np.arange(start, end + step, step)]
            elif num_variables == 2:
                results = [(x, y, multi_variable_function(x, y, function_expression)) for x in np.arange(start, end + step, step) for y in np.arange(start, end + step, step)]

            output_file_name = os.path.join(output_folder, f'lab1fun{i}dzi{j} {start} {end} {step:.3f}.dat')
            save_results(output_file_name, results, num_variables)

if __name__ == "__main__":
    x_range_start = 1  # Starting range
    x_range_end = 6    # Ending range
    file_prefix = 'lab1fun'

    process_data(file_prefix, x_range_start, x_range_end)
    print("Wyniki zostały zapisane do plików")
```

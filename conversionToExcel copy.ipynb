```python
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plotowanie_jednoargumentowe(p, k, funkcja):
    y_axis = []
    x_axis = []
    krok = 0.1
    i = int(p)
    koniec = int(k)
    
    while i < koniec:
        X1 = i
        if i == 0:
            X1 = 0.00000001
        y_axis.append(eval(funkcja))
        x_axis.append(i)
        i += krok
    
    return x_axis, y_axis

def wyznaczenie_przedzialow(nazwa_pliku):
    tabela = nazwa_pliku[:-4].split()
    if tabela[-1] == "pi":
        return -3.14, 3.14
    return tabela[-2], tabela[-1]

def wypisz_funkcje_jednoargumentowa(nazwa_pliku, sciezka_docelowa):
    with open(nazwa_pliku, 'r') as plik:
        zawartosc = plik.read()
    if len(zawartosc) > 0:
        poczatek, koniec = wyznaczenie_przedzialow(nazwa_pliku)
        x, y = plotowanie_jednoargumentowe(poczatek, koniec, zawartosc)
        plt.grid(True)
        plt.plot(x, y)
        nazwa_obrazka = f"img{nazwa_pliku[nazwa_pliku.rfind(os.path.sep)+5:-4]}.png"
        plt.savefig(os.path.join(sciezka_docelowa, nazwa_obrazka))
        plt.show()
        return os.path.join(sciezka_docelowa, nazwa_obrazka)

def plotowanie_dwuargumentowe(p, k, funkcja):
    y_axis = []
    x1_axis = []
    x2_axis = []
    krok = 0.1
    i = int(p)
    koniec = int(k)
    
    while i < koniec:
        j = int(p)
        X1 = i
        while j < koniec:
            X2 = j
            y_axis.append(eval(funkcja))
            x1_axis.append(i)
            x2_axis.append(j)
            j += krok
        i += krok
    
    return x1_axis, x2_axis, y_axis

def wypisz_funkcje_dwuargumentowa(nazwa_pliku, sciezka_docelowa):
    with open(nazwa_pliku, 'r') as plik:
        zawartosc = plik.read()
    if len(zawartosc) > 0:
        poczatek, koniec = wyznaczenie_przedzialow(nazwa_pliku)
        x1, x2, y = plotowanie_dwuargumentowe(poczatek, koniec, zawartosc)
        
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_trisurf(x1, x2, y, cmap='viridis')
        
        ax.set_xlabel('X1')
        ax.set_ylabel('X2')
        ax.set_zlabel('Y')
        
        nazwa_obrazka = f"img{nazwa_pliku[nazwa_pliku.rfind(os.path.sep)+5:-4]}.png"
        plt.savefig(os.path.join(sciezka_docelowa, nazwa_obrazka))
        plt.show()
        
        return os.path.join(sciezka_docelowa, nazwa_obrazka)

def znajdz_pliki_done(folder, pattern):
    pliki_done = []
    folder_done = os.path.join(folder, "files", "done")
    folder_img = os.path.join(folder, "files", "img")
    
    for plik in os.listdir(folder_done):
        if plik.startswith(pattern) and not plik.endswith("xmxp2.dat"):
            pelna_sciezka = os.path.join(folder_done, plik)
            nazwa = wypisz_funkcje_jednoargumentowa(pelna_sciezka, folder_img)
            print(nazwa)
            
    return pliki_done
```

```python
folder_glowny = os.getcwd()  # ścieżka do twojego folderu
pliki_done = znajdz_pliki_done(folder_glowny, "done zad1 fun5")

for plik in pliki_done:
        # wypisz_funkcje_jednoargumentowa(plik)
        wypisz_funkcje_dwuargumentowa(plik)
```
